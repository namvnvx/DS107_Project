```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/namvnvx/DS107_Project/main/Demo_data/data_covid.csv')
```

```python
df.head()
```

```python
list_na_col = [fea for fea in df.columns if df[fea].isnull().sum() > 1]
for fea in list_na_col:
	print(f'{fea} {df[fea].isnull().mean():.3f} %  --- {df[fea].isnull().sum()} mising value')
```

```python
data = pd.DataFrame(data=df)
```

```python
# Tách dữ liệu thành 2 phần
tw_data = data['clean_tweet'].values
cls_data = data['class'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(tw_data, cls_data, test_size=0.2, random_state=42)
```

```python
x1 = pd.DataFrame(data = X_train, columns=['tweet'])
y1 = pd.DataFrame(data = y_train, columns=['class'])

df1 = pd.concat([x1, y1], axis = 1)
df1.to_csv('train_data.csv', index=False, encoding='utf-8')
```

```python
df1
```

```python
x2 = pd.DataFrame(data = X_test, columns=['tweet'])
y2 = pd.DataFrame(data = y_test, columns=['class'])

df2 = pd.concat([x2, y2], axis = 1)
df2.to_csv('test_data.csv', index=False, encoding='utf-8')
```
